## Load Dependencies

```
import pandas as pd
import seaborn as sns
from matplotlib import pyplot
```

## Create chart

```
df = pd.read_csv('https://github.com/mcgovey/compvision-playing-card-detection/raw/master/ref/model_results.csv', header=None)
df.columns =['Card','Precision']
```

```

a4_dims = (20, 15)
fig, ax = pyplot.subplots(figsize=a4_dims)
my_plt = sns.barplot(ax=ax, x="Precision", y="Card", data=df, palette="Blues_d")
```
